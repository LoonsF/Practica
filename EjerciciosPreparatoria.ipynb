Ejercicio 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('ImagenCompleta.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

recorte = img[600:800, 650:840]  
recorte_resized = cv2.resize(recorte, (img.shape[1], img.shape[0]))

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis("on")
plt.title("Imagen Original")

plt.subplot(1, 2, 2)
plt.imshow(recorte_resized)
plt.axis("on")
plt.title("Imagen Recortada")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h, w = 250, 300 
degradado = np.linspace(0, 1, h, dtype=np.float32).reshape(h,1)

plt.figure(figsize=(5, 5))
plt.imshow(degradado, cmap='viridis', aspect='auto') 
plt.axis("off")
plt.title("Degradado Lineal")

plt.show()
```

Ejercicio 3

```python
import numpy as np
import matplotlib.pyplot as plt

h, w = 8, 8 
tamao_casilla = 100

tablero = np.zeros((h * tamao_casilla, w * tamao_casilla, 3), dtype=np.uint8)

for i in range(h):
    for j in range(w):
        if (i + j) % 2 == 0:
            tablero[i*tamao_casilla:(i+1)*tamao_casilla, j*tamao_casilla:(j+1)*tamao_casilla] = (255, 255, 255)
plt.figure(figsize=(6, 6))
plt.imshow(tablero)
plt.axis("on")
plt.title("Tablero de Ajedrez")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('ImagenCompleta.jpg', cv2.IMREAD_GRAYSCALE)

h, w = img.shape
degradado = np.linspace(0.2, 0, h, dtype=np.float32).reshape(h, 1)
img_deg = (img * degradado).astype(np.uint8)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis("on")
plt.title("Imagen Original")

plt.subplot(1, 2, 2)
plt.imshow(img_deg, cmap='gray')
plt.axis("on")
plt.title("Imagen con Degradado Lineal")
plt.show()
```
